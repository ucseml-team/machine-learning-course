# Jupyter

* Project Jupyter was born out of the IPython Project in 2014 as it evolved to support interactive data science and scientific computing across all programming languages.

* El notebook es un interprete interactivo en la web, pero que permite mezclar código, videos, imágenes, markdown, latex y gráficas...

* Un notebook es una sucesión de celdas, donde cada una puede ser código, texto, etc

### Labs vs Notebooks

* Un "notebook" es un documento en particular (similar a decir "un powerpoint", o "un word", etc).

* Se pueden abrir corriendo en la consola "jupyter notebook", y luego eligiendo en la web que se abre, el notebook correspondiente.

* O se pueden abrir corriendo en la consola "jupyter lab", y usando el IDE que se abre en la web para trabajar con muchos notebooks y más.

* Esto que estás viendo es un notebook!

### Uso

1. Abrir una consola
2. Moverse al directorio donde tienen los notebooks con los que van a trabajar
3. Correr `uv run jupyter notebook`, o `uv run jupyter lab`, dependiendo de lo que quieran usar.

# Pandas


* Permite trabajar con tablas de datos

* Muy rápido

* No tan intuitivo a veces, no tengan miedo de mirar la doc, tiene buenos ejemplos

* Se basa en dos estructuras de datos básicas: Series y DataFrames.

### Series vs DataFrames

* Una Serie es una secuencia de datos. 

* Cada dato tiene un índice dentro de la serie.

* Un DataFrame es una tabla de datos (que también tiene un índice).

* Cada columna de un DataFrame es en realidad una Serie.

### Filtrado de datos

```python
alice_passes = passes[passes.nickname == "alice"]

alice_passes
```

```python
first_half_passes = passes[passes.start.dt.month < 7]

first_half_passes
```

```python
first_half_passes.sample(10)
```

```python
today = datetime.utcnow()

when_to_see_alice = passes[
    (passes.nickname == "alice")
    & (passes.elevation > 25)
    & passes.start.between(today, 
                           today + timedelta(days=2))
    & (passes.start.dt.hour >= 9)
]

when_to_see_alice
```

```python
when_to_see_alice.iloc[0]
```

```python
when_to_see_alice.iloc[0].elevation
```

### Operar columnas

```python
durations = passes.end - passes.start

durations
```

```python
durations.describe()
```

```python
passes['duration'] = durations
```

```python
passes.sample(5)
```

```python
del passes['center']
del passes['end']
```

```python
passes.sample(5)
```

```python
passes[passes.duration > timedelta(minutes=11)]
```

```python
passes['duration_seconds'] = passes.duration.dt.total_seconds()

passes.sample(5)
```

```python
passes['is_good'] = (passes.duration_seconds > 600) & (passes.elevation > 30)

passes.sample(10)
```

```python
passes.is_good.value_counts()
```

```python
passes[passes.nickname == "alice"].is_good.value_counts()
```

### Agrupamiento

```python
passes.groupby("nickname").aggregate({"elevation": "mean", "duration": "max", "start": "count"})
```

### Joins

```python
from_satellogic = pd.DataFrame(
    [
        ("zarya", False),
        ("alice", True),
        ("amelia", True),
        ("joan", True),
        ("nancy", True),
    ], 
    columns=["nickname", "is_satellogic"],
)

from_satellogic
```

```python
from_satellogic = from_satellogic.set_index("nickname")

from_satellogic
```

```python
passes = passes.join(from_satellogic, on="nickname")
```

```python
passes.sample(5)
```

```python
(passes.is_good & passes.is_satellogic).value_counts()
```

# Graficar

### Alternativas

* Pandas sabe generar gráficos de sus datos (usa Matplotlib por dentro)

* Podemos usar Matplotlib directamente para tener más control

* Podemos usar Plotly para generar gráficos más lindos y con código un poco más lindo

* Podemos usar Seaborn para generar gráficos un poco menos lindos, pero que son más compatibles con matplotlib

### Desde pandas

```python
passes.nickname.value_counts().plot.bar()
```

```python
passes.plot.scatter(x="elevation", y="duration_seconds")
```

```python
passes[passes.nickname == "alice"].plot.scatter(x="elevation", y="duration_seconds")
```

### Directo con Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(x=passes.elevation, y=passes.duration_seconds)
```

```python
plt.scatter(x=passes.elevation, y=passes.duration_seconds)
plt.xlabel("elevation in degrees")
plt.ylabel("duration in seconds")
```

```python
for sate_nick in passes.nickname.unique(): 
    sate_passes = passes[passes.nickname == sate_nick]
    plt.scatter(
        x=sate_passes.elevation, 
        y=sate_passes.duration_seconds, 
        label=sate_nick,
    )

plt.title("elevation vs duration")
plt.xlabel("elevation in degrees")
plt.ylabel("duration in seconds")
plt.legend()
```

### Gráficos más lindos con plotly

```python
import plotly.express as px
```

```python
px.scatter(passes, x="elevation", y="duration_seconds", color="nickname", title="Elevation vs duration", 
           height=600)
```

```python
px.histogram(passes, x="nickname", height=600)
```

```python
px.histogram(passes, x="nickname", color="is_good", height=600)
```

```python
px.histogram(passes, x="nickname", color="is_good", barmode="group", height=600)
```

Dato importante: esto por default no funciona en Jupyter Lab (sí en Jupyter Notebook). Para que funcione en Juputer Lab hay que seguir los pasos explicados en la [doc oficial](https://plotly.com/python/getting-started/#jupyterlab-support-python-35).
